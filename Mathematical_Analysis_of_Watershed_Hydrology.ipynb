### 1. Introduction

Understanding how rainfall transforms into river discharge is a fundamental problem in hydrology and environmental engineering. When precipitation falls over a watershed, part of the water infiltrates into the soil, part is stored temporarily, and the remaining portion becomes surface runoff that contributes to streamflow. The shape of the resulting hydrograph reflects the physical characteristics of the landscape.

Forest cover plays a crucial role in this process. Vegetation increases infiltration, enhances surface roughness, and delays runoff response. When forest density decreases, water moves more rapidly across the surface, often leading to higher peak discharge and greater flood risk.

Because these processes evolve continuously in time, they can be described using mathematical functions. Calculus provides the tools needed to:

 * Describe how discharge changes over time;
 * Determine when peak flow occurs;
 * Compute the total water volume transported;
 * Analyze how sensitive the system is to changes in forest cover.

This project develops a simplified mathematical model of watershed runoff in order to quantitatively evaluate how different levels of forest cover influence hydrograph dynamics.

#### Why This Calculation Is Necessary

Flood prediction, watershed management, and infrastructure design require quantitative estimates of peak discharge and total runoff volume. Without mathematical modeling, it is impossible to:

 * Compare scenarios with different land cover conditions;
 * Estimate the effect of deforestation on flood intensity,
 * Evaluate the retention capacity of a watershed.

By translating hydrological processes into mathematical expressions, we gain predictive capability. Even a simplified model allows us to isolate the influence of key parameters—such as runoff coefficient and storage rate—and understand their impact on system behavior.

The goal is not to reproduce every physical mechanism in detail, but to construct a mathematically transparent framework that reveals cause-and-effect relationships between forest cover and runoff response.

### 2. Theoretical Background
#### Watershed and Runoff Processes

A watershed (drainage basin) is a geographical area in which all precipitation drains toward a common outlet, such as a river or stream.
When rainfall occurs over a watershed, water is partitioned into three primary components:

 * Infiltration into the soil
 * Temporary storage (surface depressions, soil moisture, vegetation)
 * Surface runoff contributing to streamflow

The balance between these components depends on land cover, soil type, slope, and rainfall intensity.
Forest cover significantly influences this balance. Vegetation:

 * Increases infiltration capacity
 * Enhances evapotranspiration
 * Reduces surface flow velocity
 * Increases surface roughness

As a result, forested watersheds generally exhibit delayed runoff response and lower peak discharge compared to deforested or urbanized areas.

### 3. Mathematical Model

To analyze how forest cover affects surface runoff during rainfall events, a simplified mathematical model is constructed. The model combines classical hydrological concepts with calculus in order to analyze how discharge changes over time.
The modeling process consists of several steps:
 * Extending the Rational Method into a time-dependent form
 * Defining a mathematical model for rainfall intensity
 * Deriving a discharge function
 * Using the first derivative to determine peak discharge
 * Using the second derivative to analyze hydrograph curvature
 * Using integration to calculate total runoff volume

This approach allows the hydrological process to be studied using standard tools from Calculus and Hydrology.

#### Time-Dependent Extension of the Rational Method

A commonly used hydrological equation for estimating peak runoff is the Rational Method.
The classical formula is:

$$Q = C \cdot I \cdot A$$
where:
 * $Q$ - peack discharge (m<sup>3</sup>/s), how much water leaves a watershed during a storm
 * $C$ - runoff coefficient (dimensionless), represents how much rainfall becomes runoff
 * $I$ - rainfall intensity (mm/h), represents storm intensity
 * $A$ - watershed area (km<sup>2</sup>), represents the size of the watershed

However, the classical Rational Method provides only one value (peak discharge) and does not describe how runoff changes during the storm.
To analyze the process mathematically, rainfall intensity must be treated as a function of time or rainfall intensity at time *t*:

$$I = I(t)$$


This leads to a time-dependent discharge model:

$$Q(t) = C \cdot I(t) \cdot A$$

where

$Q(t)$ – discharge at time *t*

This transformation allows us to analyze the runoff using calculus.

##### Rainfall Intensity Model (Gaussian Function)
Rainfall during storms often follows a bell-shaped pattern with weak rainfall at the beginning, maximum intensity in the middle and decreasing intensity afterward. Thats why this type of distribution can be modeled using the **Gaussian function**, which is widely used in statistics and physics. The Gaussian rainfall model is:
$$ I(t) = I_{max}e^{-\frac{(t - \mu)^2}{2\sigma^2}} $$

where

 * $Imax$ – maximum rainfall intensity,
 * $\mu$ - time of peak rainfall,
 * $\sigma$ - spread of the rainfall event.

The interpretaion of the formula is: The Gasussian function creates a realistic rainfall pattern. Before $\mu$ rainfall increases, at $\mu$ rainfall reaches maximum and after $\mu$ rainfall gradually decreases. This produces a realistic storm profile that can be used to simulate runoff.

##### Discharge Function Derivation
The discharge function is obtained by substituting the rainfall intensity model into the time-dependent Rational Method.

Starting equation:
$$ Q(t) = C \cdot A \cdot I(t)$$
Substituting the Gaussian rainfall model where the function represents the runoff hydrograph of the watershed:
$$ Q(t) = C \cdot A \cdot I_{max}e^{-\frac{(t - \mu)^2}{2\sigma^2}} $$

##### Physical meaning

The parameters influence the hydrograph as follows:
 * $C$ represents forest cover influence
 * $A$ is watershed size
 * $I_{max}$ is storm intensity
 * $\mu$ time of peak rainfall
 * $\sigma$ duration of rainfall event

For example when we have dense forest $C$ is smaller, but when we have deforested land $C$ is larger.
Thus forest density directly affects the magnitude of runoff.

#### Peak Discharge (First Derivative Analysis)

To determine when the maximum discharge occurs, we compute **the first derivative** of the discharge function.

$$\frac{dQ}{dt}$$

The derivative measures the **rate of change of discharge with respect to time.**

##### Critical point condition

Peak discharge occurs when
$$\frac{dQ}{dt} = 0$$
At this point the hydrograph stops increasing and begins to decrease. The interpretation is: before the peak → discharge increases, at the peak → maximum runoff, after the peak → discharge decreases.
This step allows the model to mathematically identify the **maximum flood discharge.**

##### Curvature and Hydrograph Shape (Second Derivative)

The shape of the hydrograph can be analyzed using the second derivative.
$$\frac{d^2Q}{dt^2}$$
The interpretation is: The second derivative describes concavity and convexity of the curve. If is **bigger** then zero the graph is **convex**, if is **smaller** then zero the graph is **concave.**

##### Hydrological meaning

This helps interpret watershed behavior. Hydrologically, the curvature of the hydrograph provides insight into how the watershed responds to rainfall over time. During the early stage of a storm event, rainfall begins to accumulate on the surface and infiltration processes are still active. As more water reaches the ground and the soil gradually becomes saturated, surface runoff begins to increase more rapidly. This causes the rising limb of the hydrograph to become progressively steeper, indicating an acceleration in the rate at which water is entering the stream system.

As the storm progresses and approaches its peak intensity, the watershed reaches a point where the increase in discharge slows down. After the peak discharge is reached, rainfall intensity typically decreases and the available water in the system begins to drain. At this stage, the rate of runoff gradually declines, causing the hydrograph to flatten and transition into the recession limb.

Therefore, the curvature of the hydrograph reflects the changing physical conditions within the watershed. In particular, it indicates how soil saturation, infiltration capacity, and temporary water storage influence the acceleration and deceleration of runoff during a storm event.

#### Total Runoff Volume (Definite Integral)

The total water volume passing through the watershed during the storm is calculated using a definite integral.

$$ V = \int_{t_1}^{t_2} Q(t)\; dt$$

where:
 * $V$ - total runoff volume
 * $t_1$, $t_2$ - beginning and end of the rainfall event

The definite integral of the discharge function represents the total amount of water that passes through the river channel during a storm event. Graphically, this integral corresponds to the area under the hydrograph curve over a given time interval. Since the hydrograph shows how discharge changes with time, integrating this function allows us to accumulate all the instantaneous discharge values and determine the total runoff volume produced by the watershed.

If the area under the hydrograph is large, it means that the river maintained relatively high discharge values for a longer period of time, which results in a greater total volume of water moving through the system. Conversely, if the area under the curve is smaller, the discharge was either lower in magnitude, shorter in duration, or both. In this case, the watershed produces a smaller runoff volume. This interpretation is particularly useful when comparing different land cover scenarios, as changes in forest density can influence the magnitude and duration of runoff, and therefore the total water volume represented by the hydrograph.

Forests have a significant influence on the way rainfall is transformed into surface runoff. Vegetation intercepts part of the precipitation in the canopy, where a portion of the water can evaporate before reaching the ground. At the same time, forest soils typically have a well-developed structure due to root systems and organic matter, which improves their ability to absorb and store water. This leads to higher infiltration rates and reduces the amount of water that immediately becomes surface runoff.

As a result, watersheds with dense forest cover generally have a lower runoff coefficient $C$. A smaller value of $C$ means that a smaller fraction of the rainfall contributes to direct runoff. In terms of the hydrograph, this leads to a smaller area under the curve and therefore a lower total runoff volume compared to watersheds with reduced forest cover.
